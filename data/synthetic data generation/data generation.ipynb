```python
import openai
import pandas as pd
import time
import random
from datetime import datetime

# Set your OpenAI API key
openai.api_key = ".........."  # Replace with your valid OpenAI key

# Define a general prompt to generate diverse health insurance Q&A samples
general_prompt = (
    "Generate a unique and realistic general health insurance Q&A. "
    "The topic can be anything relevant to health insurance including coverage, deductibles, claims, plan options, network hospitals, telemedicine, renewals, and government schemes. "
    "Write the result in the format:\nQ: <question>\nA: <answer>"
)

# Store generated samples here
synthetic_data = []

# Total number of samples to generate
num_samples = 300

# Start timer
start_time = datetime.now()
print(f"Generation started at: {start_time.strftime('%Y-%m-%d %H:%M:%S')}")

# Generate Q&A samples one by one
for i in range(num_samples):
    try:
        # Track time for each sample
        sample_start_time = time.time()

        # Call OpenAI API to generate the Q&A
        response = openai.completions.create(
            model="gpt-4o-mini",
            prompt=general_prompt,
            temperature=0.9,
            max_tokens=400
        )

        # Get the generated output
        output = response.choices[0].text.strip()

        # Check and extract the question and answer
        if "Q:" in output and "A:" in output:
            question = output.split("Q:")[1].split("A:")[0].strip()
            answer = output.split("A:")[1].strip()
            synthetic_data.append({"question": question, "answer": answer})
            print(f"Sample {i+1} generated successfully")
        else:
            print(f"Invalid format at sample {i+1}")

        # Sleep a bit to avoid rate limit
        time.sleep(random.uniform(0.5, 1.2))

    except Exception as e:
        print(f"Error at sample {i+1}: {e}")
        time.sleep(5)

# Save the synthetic dataset
df = pd.DataFrame(synthetic_data)
df.to_csv("general_synthetic_health_insurance_qa_1.csv", index=False)

# End timer
end_time = datetime.now()
elapsed_time = end_time - start_time

print(f"Generated {len(synthetic_data)} Q&A samples successfully.")
print(f"Finished at: {end_time.strftime('%Y-%m-%d %H:%M:%S')}")
print(f"Total time: {elapsed_time}")
```

```python
import openai
import pandas as pd
import time
import random
from datetime import datetime

# Set your OpenAI API key
openai.api_key = ""  # Replace with your actual API key

# Load the cleaned real-world health insurance dataset
df = pd.read_csv("/content/health_insurance_real_world_clean.csv")

# Define categories and their corresponding prompts
areas = {
    "Policy Information": "Generate a general health insurance Q&A about policy details, including policy terms, renewal procedures, and how to understand what your insurance policy covers.",
    "Claims and Reimbursement": "Generate a general health insurance Q&A about filing insurance claims, how the reimbursement process works, and common issues that can arise during claims processing.",
    "Customer Support": "Generate a general health insurance Q&A about customer support, including how to contact support, what kinds of help are available, and common support issues.",
    "Plan Recommendations": "Generate a general health insurance Q&A about how to choose the right health insurance plan based on age, income, family size, and medical needs.",
    "Premium Payment Management": "Generate a general health insurance Q&A about managing premium payments, including payment options, auto-pay, due dates, and what happens if a payment is missed.",
    "Appointment Booking": "Generate a general health insurance Q&A about how to book doctor appointments through your insurance, including referrals, networks, and using insurance portals.",
    "Coverage Checker": "Generate a general health insurance Q&A about how to check if a treatment, medication, or provider is covered under your insurance plan.",
    "Eligibility and Enrollment": "Generate a general health insurance Q&A about eligibility for health insurance, how to enroll, and when you can apply (e.g., open enrollment, special enrollment periods).",
    "Health Insurance Plans": "Generate a general health insurance Q&A about different types of health insurance plans (e.g., HMO, PPO, EPO), their benefits, and how to choose the right plan.",
    "Cost of Health Insurance": "Generate a general health insurance Q&A about how the cost of health insurance is determined, including premiums, deductibles, copayments, and out-of-pocket maximums."
}

# Create a list to store the generated Q&A data
synthetic_data = []

# Set the number of Q&A samples to generate for each category
samples_per_category = 250

# Track total generated samples and start time
total_generated_samples = 0
start_time = datetime.now()
print(f"Generation started at: {start_time.strftime('%Y-%m-%d %H:%M:%S')}")

# Loop through each category and generate the specified number of samples
for category, base_prompt in areas.items():
    for i in range(samples_per_category):
        try:
            # Record the time before starting this sample
            sample_start_time = time.time()

            # Generate a Q&A using the OpenAI API
            response = openai.completions.create(
                model="gpt-4o-mini",
                prompt=f"{base_prompt} Provide the question and a detailed answer. Format the response as 'Q: <question> A: <answer>'",
                temperature=0.85,
                max_tokens=400
            )

            # Extract the generated text
            output = response.choices[0].text.strip()

            # Check if the output format is correct and extract question and answer
            if "Q:" in output and "A:" in output:
                q = output.split("Q:")[1].split("A:")[0].strip()
                a = output.split("A:")[1].strip()
                synthetic_data.append({"question": q, "answer": a, "category": category})
                total_generated_samples += 1
            else:
                print(f"Invalid format in response {i} for {category}")

            # Measure how long this sample took
            sample_end_time = time.time()
            sample_generation_time = sample_end_time - sample_start_time
            print(f"Generated Sample {i+1} in {sample_generation_time:.2f} seconds for {category}")

            # Sleep a bit to avoid hitting the rate limit
            time.sleep(random.uniform(0.5, 1.5))

        except Exception as e:
            print(f"Error in {category} - Sample {i}: {e}")
            time.sleep(5)

# Convert the generated data into a DataFrame
synthetic_df = pd.DataFrame(synthetic_data)

# Save the synthetic Q&A dataset to a CSV file
synthetic_df.to_csv("synthetic_health_insurance_qna_2.csv", index=False)

# Show end time and total elapsed time
end_time = datetime.now()
elapsed_time = end_time - start_time
print(f"Successfully generated {total_generated_samples} synthetic Q&As.")
print(f"Generation completed at: {end_time.strftime('%Y-%m-%d %H:%M:%S')}")
print(f"Total generation time: {elapsed_time}")
```
